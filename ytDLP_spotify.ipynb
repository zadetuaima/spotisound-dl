```python
import spotipy
import pandas as pd
from spotipy.oauth2 import SpotifyOAuth
import json
import yt_dlp
import os

key = "key.json"
with open(key, "r") as file:
    api_tokens = json.load(file)

client_secret = api_tokens['client_secret']
client_id = api_tokens['client_id']
redirectURI = api_tokens['redirect']
username = api_tokens['username']

playlist_id = 'https://open.spotify.com/playlist/374IfZMn0uLtxM64bwMnEB?si=c7a4b8bb620844f5' 
download_path = 'C:/Users/Zade/Downloads'  

sp = spotipy.Spotify(auth_manager=SpotifyOAuth(client_id=client_id,
                                               client_secret=client_secret,
                                               redirect_uri=redirectURI,
                                               scope="playlist-read-private"))
```

```python
def download_from_youtube(query, download_path):
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '320',
        }],
        'outtmpl': f'{download_path}/%(title)s.%(ext)s',
        'quiet': True
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        try:
            print(f"Downloading {query}...")
            ydl.download([f"ytsearch:{query}"])
        except Exception as e:
            print(f"Error downloading {query}: {e}")
```

```python
def get_playlist_tracks(playlist_id):

    results = sp.playlist_tracks(playlist_id)
    tracks = results['items']  
    return tracks
```

```python
print(get_playlist_tracks(playlist_id))
```

```python
tracks = []

for item in get_playlist_tracks(playlist_id):
    track = item['track']
    artist_name = track['artists'][0]['name']
    track_name = track['name']
    search_query = f"{artist_name} - {track_name}"
    tracks.append((search_query, track_name))  # Store query and track name in a tuple

    print(search_query)
```
